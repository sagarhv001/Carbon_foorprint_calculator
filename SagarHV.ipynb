```python
import re

def remove_emoji(string):
    if isinstance(string, str):
        emoji_pattern = re.compile("["
            u"\U0001F600-\U0001F64F"  # emoticons
            u"\U0001F300-\U0001F5FF"  # symbols & pictographs
            u"\U0001F680-\U0001F6FF"  # transport & map symbols
            u"\U0001F700-\U0001F77F"  # alchemical symbols
            u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
            u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
            u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
            u"\U0001FA00-\U0001FA6F"  # Chess Symbols
            u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
            u"\U0001F000-\U0001F0FF"  # Mahjong Tiles
            u"\U0001F200-\U0001F2FF"  # Enclosed Ideographic Supplement
            u"\U0001F400-\U0001F4FF"  # Miscellaneous Symbols and Pictographs
            u"\U0001F500-\U0001F5FF"  # Miscellaneous Symbols and Pictographs
            u"\U0001F600-\U0001F64F"  # Emoticons
            u"\U0001F680-\U0001F6FF"  # Transport & Map Symbols
            u"\U0001F700-\U0001F77F"  # Alchemical Symbols
            u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
            u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
            u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
            u"\U0001FA00-\U0001FA6F"  # Chess Symbols
            u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
            u"\U0001F000-\U0001F0FF"  # Mahjong Tiles
            u"\U0001F200-\U0001F2FF"  # Enclosed Ideographic Supplement
            u"\U0001F400-\U0001F4FF"  # Miscellaneous Symbols and Pictographs
            u"\U0001F500-\U0001F5FF"  # Miscellaneous Symbols and Pictographs
            u"\U0001F600-\U0001F64F"  # Emoticons
            u"\U0001F680-\U0001F6FF"  # Transport & Map Symbols
            u"\U0001F700-\U0001F77F"  # Alchemical Symbols
            u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
            u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
            u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
            u"\U0001FA00-\U0001FA6F"  # Chess Symbols
            u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
            u"\U0001F000-\U0001F0FF"  # Mahjong Tiles
            u"\U0001F200-\U0001F2FF"  # Enclosed Ideographic Supplement
            u"\U0001F400-\U0001F4FF"  # Miscellaneous Symbols and Pictographs
            u"\U0001F500-\U0001F5FF"  # Miscellaneous Symbols and Pictographs
            "]+", flags=re.UNICODE)
        return emoji_pattern.sub(r'', string)
    else:
        return string

def remove_special_characters(string):

    allowed_words = ["a**hole", "c*nt", "b*tch"]
    for word in allowed_words:
        string = string.replace(word, word.replace('*', ''))
    string = re.sub(r'[^a-zA-Z\s]+', '', string)
    return string


df1['text'] = df1['text'].apply(remove_emoji)
df1['text'] = df1['text'].apply(remove_special_characters)

df2['text'] = df2['text'].apply(remove_emoji)
df2['text'] = df2['text'].apply(remove_special_characters)
```

```python


def split_text_to_rows(df, text_column):
    new_rows = []
    for _, row in df.iterrows():
        # Check if the value in the text_column is a string
        if isinstance(row[text_column], str):
            texts = row[text_column].split('\n')
            for text in texts:
                new_row = row.copy()
                new_row[text_column] = text.strip()
                new_rows.append(new_row)
        else:
            new_rows.append(row)  # If not a string, keep the row as it is
    new_df = pd.DataFrame(new_rows)
    return new_df
df1 = split_text_to_rows(df1, 'text')
df1
```

```python


from sklearn.feature_extraction.text import TfidfVectorizer

# Instantiate the TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=500)

# Fit and transform the text data
text_tfidf = tfidf_vectorizer.fit_transform(combined_df['text'])

# Convert the TF-IDF matrix into a DataFrame
text_tfidf_df = pd.DataFrame(text_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Combine the TF-IDF DataFrame with the original DataFrame
combined_df = pd.concat([combined_df.reset_index(drop=True), text_tfidf_df], axis=1)

# Print the combined DataFrame
print(combined_df)
```
